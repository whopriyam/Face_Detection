```python
import cv2
#Creat a cascade classifier object
face_cascade = cv2.CascadeClassifier("/Users/priyambasu/desktop/haarcascade_frontalface_default.xml")
#Reading image as it is
img = cv2.imread("/Users/priyambasu/desktopReal_madrid", 1)
#Reading image as gray scale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#Search co-ordinates of image
faces = face_cascade.detectMultiScale(gray_img, scaleFactor = 1.05, minNeighbors = 5)
print(type(faces))
print(faces)

for x,y,w,h in faces:
    img = cv2.rectangle(img, (x,y), (x+w, y+h), (0, 255, 0), 3)
    
cv2.imshow("Gray", img)

cv2.waitKey(0)

cv2.destroyAllWindows()
```
